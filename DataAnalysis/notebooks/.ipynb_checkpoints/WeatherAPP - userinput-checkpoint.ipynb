# Real-Time Weather Dashboard with OpenWeather API

The project builds a Weather Dashboard using the OpenWeather API to fetch and display real-time weather details for one or two specified locations. Users can compare key weather data such as temperature, humidity, wind speed, and air conditions for these locations, with an option to save the data to a CSV file for future analysis. 

The dashboard aims to enhance skills in API integration, interactive interface, and data visualization (via Matplotlib).
